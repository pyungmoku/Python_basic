# 로컬 변수와 글로벌 변수

````python
EPOCH = 2
BATCH = 5 # 글로벌 변수

def train(weight): # weight는 매개변수
    x = 1 # x는 로컬 변수
    for i in range(EPOCH): #글로벌 변수는 안에서도 쓸수 있다.
        for j in range(BATCH):
            print(f"{i} 번째 Epoch이며 {j} 번째 batch 입니다.")
            weight += 1
    return weight

train(0)
````

````python
class Robot:
    def __init__(self, height, sentence): # 밖에서 넘어오는 매개변수는 무조건 self 뒤에
        self.height = height
        self.sentence = sentence
    def greeting(self):
        print(self.sentence)
        print(f"제 키는 {self.height} 이에요")

bot1 = Robot(180, "안녕")
bot1.greeting()

print(bot1.height)
print(bot1.sentence)
bot1.height -=10 
bot1.greeting()

bot2 = Robot()
````

````python
class Robot:
    def __init__(self, height, sentence): # 밖에서 넘어오는 매개변수는 무조건 self 뒤에
        self.__height = height # 비공개로 만들고 싶은 속성은 앞에 밑줄 두개
        self.sentence = sentence
    def greeting(self):
        print(self.sentence)
        self.__height -=10
        print(f"제 키는 {self.__height} 이에요")
bot1 = Robot(180, "hi")
bot1.greeting()
# self.height는 밖에서 접근이 안된다. 내부에서는 괜찮다.
````

````python
# 상속은 효율화다
class Robot: #부모 클래스
    def greeting(self):
        print("안녕하세요")

class Study_bot(Robot): # 자식 클래스, Robot 클래스를 상속받음 -> Robot이 가지는 함수 사용 가능
    def study(self):
        print("열심열심")

bot = Study_bot()
bot.greeting()
bot.study()
````

````python
# 메서드 오버라이딩(overriding) 1
class Robot: 
    def greeting(self):
        print("안녕하세요")

class Study_bot(Robot): 
    def greeting(self): #부모의 greeting을 물려받았으나, 여기서 다시 정의 하면 덮어써버림
        print("공부하는 로봇입니다")

bot1 = Study_bot()
bot1.greeting()
````

````python
# 메서드 오버라이딩(overriding) 2
class Robot: 
    def greeting(self):
        print("안녕하세요 로봇입니다.")

class Study_bot(Robot): 
    def greeting(self): 
        super().greeting() # super()가 부모 클래스를 가리킴 #부모의 것을 이렇게 사용하는 것 가능
        print("공부하는 로봇입니다.")# 부모의 원래 기능은 유지ㅡ 거기에 덧붙이고 싶을 때는 오버라이딩 사용 가능

bot1 = Study_bot()
bot1.greeting()
````

````python
class Robot:
    def func1(self):
        print("부모 클래스의 func1 실행")
        self.money = 100
class Study_bot(Robot):
    def func2(self):
        print("자식 클래스의 func2 실행")
        self.book = 10
bot1 = Study_bot()
bot1.func1()
bot1.func2()
print(bot1.book)
print(bot1.money) # 부모함수를 실행해야 속성(변수)를 가져올 수 있다.
````

````python
# 딥러닝 코드에선 아래와 같은 코드를 많이 보게 됨
class Robot:
    def __init__(self):
        print("부모 클래스의 init 실행")
        self.money = 100

class Study_bot(Robot):
    def __init__(self):
        super().__init__()
        print("자실 클래스의 init 실행")
        self.book = 10

bot_parent = Robot()
print(bot_parent.money)

bot1 = Study_bot()
print(bot1.book)
print(bot1.money)
````

````python
class Robot:
    def greeting(self):
        print("로봇입니다.")
class Life:
    def eat(self):
        print("냠냠")

class Study_bot(Robot, Life):
    def study(self):
        print("열심열심!")

bot1 = Study_bot()
bot1.greeting()
bot1.eat()
bot1.study()
````

````python
# 알쏭달쏭 다이아몬드 상속
class A:
    def greeting(self):
        print("A")

class B(A):
    def greeting(self):
        print("B")

class C(A):
    def greeting(self):
        print("C")

class D(B,C):# 가장 왼쪽 것(B)의 greeting을 사용하게 된다. (C,B)하면 C 를 상속한다.
    def __init__(self):
        print("D 생성")

D1 = D()
D1.greeting()
````
