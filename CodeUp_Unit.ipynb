# 기초4-1 단순 반복문

```python
# 1251 : 1 부터 100까지 출력하기

for i in range(1,101):
    print(i, end = ' ')
```

```python
# 1252 : 1 부터 n 까지 출력하기
n = int(input())
for i in range(1, n+1):
    print(i, end = ' ')
```

```python
# 1253 : a 부터 b까지 출력하기
a, b = map(int, input().split())
if a > b:
    for i in range(b, a+1):
        print(i, end = ' ')
else:
    for i in range(a, b+1):
        print(i, end = ' ')
```

```python
# 1254 : 알파벳 출력하기

a_char, b_char = input().split()
a_int = ord(a_char)
b_int = ord(b_char)

if b_int > a_int:
    for i in range(a_int, b_int+1):
        print(chr(i), end = ' ')
else: 
    for i in range(a_int, b_int-1, -1):
        print(chr(i), end = ' ')
```

```python
# 1255 : 두 실수 사이 출력하기
a, b = map(float, input().split())

interval = 0.01
n = int(round((b-a)/interval)) +1
# print(n)

for i in range(n):
    print(format((a +i*interval), ".2f"), end = ' ')
```
