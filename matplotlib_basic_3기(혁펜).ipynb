```python
# red, 별모양, dash선으로 sin 그리기
# green, 동그라미, 실선으로 cos 그리기
theta = np.linspace(0, 2*np.pi, 20)
plt.plot(theta, np.cos(theta))
plt.figure()
plt.plot(theta, np.cos(theta),"r*--", markersize=10, label = "cos")
plt.plot(theta, np.sin(theta),"go-", label = "sin")
plt.xlabel(r"$\theta$")
plt.ylabel("value")
plt.title("Cos and Sin", color="b")
plt.grid()
plt.legend()
# plt.xlim([0, np.pi])
# plt.ylim([-1.5, 1.5])
plt.axis([0, np.pi,-1.5, 1.5])
```

```python
plt.bar(range(5),[3,2,3,4,5], width=0.5)
plt.xticks(range(5), ["a", "b", "c", "d", "e"])
plt.ylim([-1, 6])
plt.xlabel("layer")
plt.ylabel("value")
```

```python
x = range(5)
y1 = [i for i in range(1,6)]
y2 = [i+100 for i in range(1,6)]

plt.figure()
plt.plot(x,y1)
plt.plot(x,y2)

plt.figure()
fig, ax1 = plt.subplots()
ax2 = ax1.twinx()
p1 = ax1.plot(x,y1,'b')
p2 = ax2.plot(x,y2,'r')
ax1.set_ylim([-10, 10]) # plt.ylim([-10,10])
ax1.set_xlabel("x")
ax1.set_ylabel("y1")
ax2.set_ylabel("y2")
ax1.set_title("y1 and y2")
ax1.grid()
plt.legend(p1+p2,["y1", "y2"], loc = "best") 
# 'best', 'upper right', 'upper left', 'lower left', 'lower right', 'right', 'center left', 'center right', 'lower center', 'upper center', 'center'
```

# 2개의 변수함수, 등고선으로 보기, 3D로 보기

```python
a = np.linspace(-1,1,10)
b = np.linspace(-1,1,10)

A, B = np.meshgrid(a,b)
print(f"A= \n {A} \n")

L = A**2 + B**2

plt.contour(A,B,L,10)
plt.xlabel("a"); plt.xlabel("b"); plt.grid()
```

```python
plt.figure(1, figsize=[12,9])
ax = plt.axes(projection="3d")
ax.view_init(elev=20, azim=-45)
ax.plot_surface(A,B,L)

plt.figure(2, figsize=[12,9])
ax = plt.axes(projection="3d")
ax.view_init(elev=20, azim=-45)
ax.plot_wireframe(A,B,L)
```

```python
from matplotlib.colors import LightSource
plt.figure(1, figsize=[12,9])
ax = plt.axes(projection="3d")
ls = LightSource(altdeg=20, azdeg=-45)
rgb = ls.shade(L, plt.cm.RdYlBu)
ax.plot_surface(A,B,L, facecolors=rgb)
```

```python
plt.figure(1, figsize=[12,9])
ax = plt.axes(projection="3d")
ax.plot_surface(A,B,L, cmap = 'viridis') # 'viridis', 'plasma', 'inferno', 'magma', 'cividis'

```

```python
plt.figure(1, figsize = [10,9])
ax = plt.axes(projection="3d")
ax.scatter(A,B,L, s=100, c="g", marker="*", edgecolor="r")

plt.figure(2, figsize = [10,9])
ax = plt.axes(projection="3d")
ax.scatter(A,B,L, s=100, c="g", marker="*", alpha=0.3)
```

```python
plt.figure(figsize = [10,9])
ax = plt.axes(projection ="3d")
ax.plot_surface(A,B,L, alpha=0.3, cmap="viridis")
ax.scatter(A, B, L, s=100, c="r", marker="*")
```

# plotly

```python
import plotly.graph_objects as go

a = np.linspace(-2,2,100)
b = np.linspace(-2,2,100)
A,B = np.meshgrid(a,b)
L = np.cos(A**2.4) + np.sin(B**2.4)

fig = go.Figure(data = [go.Surface(x=a, y=b, z=L, colorscale="viridis", opacity=0.5)])
fig.update_traces(contours_z = dict(show=True, usecolormap=True, highlightcolor="limegreen", project_z=True))
fig.update_layout(title = "cos(A^2.4) + sin(B^2.4)", width=700, height=600,
                  scene = dict(zaxis=dict(nticks=30, range=[-3,3])))
```

# 이미지 보기

```python
import matplotlib.pyplot as plt
from torchvision import datasets
test_DS = datasets.MNIST(root="data/", train=False, download= True)
```

```python
print(f"test_DS.data.shape = \n {test_DS.data.shape} \n")# (개수, 행개수, 열개수)
plt.imshow(test_DS.data[0], cmap="gray") #image show
plt.colorbar()
```

```python
for i in range(6):
    plt.subplot(2,3,i+1, xticks=[], yticks=[])
    plt.imshow(test_DS.data[i], cmap="gray")
    plt.title(test_DS.targets[i].item(), color="g")
```

```python
im = test_DS.data[0]
plt.figure(figsize = (20,15))
plt.imshow(im, cmap="gray")
plt.colorbar()

for i in range(im.shape[0]):
    for j in range(im.shape[1]):
        #plt.text(j, i,im[i,j].item())# x위치, y위치, 쓰고 싶은 글자 
        # plt.text(j, i,im[i,j].item(),
        #          horizontalalignment = "center") 
        plt.text(j, i,im[i,j].item(),
                 horizontalalignment = "center",
                color = "k" if im[i,j] > torch.max(im)/1.5 else "w" )      
plt.xticks([])
plt.yticks([])
```
