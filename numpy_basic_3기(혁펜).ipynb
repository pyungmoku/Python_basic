# 모듈과 패키지

```python
# 모듈은 변수, 함수, 클래스를 담고 있는 파일(.py)
# 패키지는 여러 모듈을 묶은 것(폴더)
# import 모듈
# import 모듈1, 모듈2
# 모듈.변수
# 모듈.함수()
# 모듈.클래스()
```

```python
import math # math라는 모듈
print(math.pi)
print(math.sqrt(4))
```

```python
import math as m
print(m.pi)
print(m.sqrt(4))
```

```python
from math import pi, sqrt
print(pi)
print(sqrt(4))
```

```python
# import 패키지
# import 패키지.모듈1
# import 패키지.모듈1, 패키지.모듈2
# 패키지.모듈.변수
# 패키지.모듈.함수()
# 패키지.모듈.클래스()
```

```python
import numpy as np # numpy는 라이브러리 (패키지와 모듈을 모아놓음 = 폴더 위에 폴더 )
a = np.array([1,2,3])
print(a)
```

```python
import numpy.random
print(numpy.random.randn(3,3)) # Normal distribution, (3,3)은 3행 3열 짜리를 의미
```

```python
import numpy.random as nr
print(nr.randn(3,3))
```

```python
from numpy.random import randn
print(randn(3,3))
```

# numpy 본격적으로

```python
import numpy as np
a = np.array([1,2,3,4]) # list가 입력으로 들어가는 함수
print(a)
print(type(a))
print(a.dtype)
print(a.shape)
b = np.array([1,2, 3.1, 4])
print(b.dtype) 
```

```python
A = np.array([[1,2],[3,4]]) # 이셀부터는 import numpy as np 실행했다고 치고 작성
print(A)
print(A.shape) # 자주 쓰임
print(A.ndim) # 차원의 수
print(A.size)
```

```python
print(np.zeros(5))
print(np.zeros_like(A))
print(np.ones(5))
print(np.zeros([3,3]))

print(np.arange(3,10,2) ) # range와 같은데 array로 만들어줌
print(np.arange(0,1, 0.1))
print(np.linspace(0,1,10))
print(np.linspace(0,1,10, endpoint=False)) # 0에서부터 1포함(default) 10개로
```

```python
a = np.array([1,2,3])
b = np.array([4,5,6])
c = a+b
print(c)
```

```python
A = np.array([[1,2,3],[1,2,3]])
B = np.array([[4,5,6],[1,1,1]])
C = A+B
D = A-B
print(C)
print(D)
print(A*B) # 요소와의 곱
print(A/B) # 요소와의 나누기
```

```python
A = np.array([[1,2],[3,4]])
B = np.array([[1,2],[3,4]])
print(A*B)
print(A@B) # 행렬 곱
```
