```python
## 001 숫자의 합 구하기 (난이도 브론즈4, 백준 11720번)
# 입력 : N개의 숫자가 공백 없이 써 있다
# 출력 : 입력으로 주어진 숫자 N개의 합

# n1 = int(input("숫자 개수"))
# n2 = list(input("공백 없이 주어진 N개의 숫자"))

n1 = int(input("")) # 숫자 개수
n2 = list(input("")) # 공백 없이 주어진 N개의 숫자

#print(f"type(n2[0])= \n {type(int(n2[0]))} \n")
sum1 = 0
for i in n2:
    sum1 += int(i)
print(f"{sum1}")

del n1, n2, sum1
```

```python
## 002 평균 구하기 (난이도 브론즈1, 백준 1546번)
# 입력1 : 시험을 본 과목의 수 N
# 입력2 : 각 과목의 원래 시험 성적(공백으로 구분하여 기입)
# 출력 : 각 과목 점수 변환(=과목 점수/최대 과목점수 *100)하여 평균


n1 = input("") # 시험을 본 과목의 수
n2 = list(map(int, input().split())) # 각 과목의 원래 시험 성적(공백으로 구분하여 기입)
# print(n2)
# print(type(n2))
score_max = max(n2)
# print(score_max)
score_sum = sum(n2)
# print(score_sum)
print(score_sum/int(n1)/score_max*100)

del n1, n2, score_max, score_sum
```

```python
## 구간 합 구하기 (난이도 실버3, 백준 11659번)
# 입력1 : 데이터의 개수(1 ~ 100,000) & 질의 개수
# 입력2 : 구간(i,j) 합을 구할 대상 배열
# 출력 : i번째 수에서 j번째 수까지의 합

# 왜..백준에서 시간초과나는지는 모르겠다.

num_data, num_question  = map(int, input().split())

data = list(map(int, input().split()))

section_sum = [0]
temp = 0
for i in data:
    temp += i
    section_sum.append(temp)

for i in range(num_question):
    start_idx, end_idx  = map(int, input().split())
    print(f"{section_sum[end_idx] - section_sum[start_idx-1]}")

del num_data, num_question, data, section_sum
```

```python
## 구간 합 구하기 (난이도 실버3, 백준 11659번)
# 입력1 : 데이터의 개수(1 ~ 100,000) & 질의 개수
# 입력2 : 구간(i,j) 합을 구할 대상 배열
# 출력 : i번째 수에서 j번째 수까지의 합
# import sys
# input = sys.stdin.readline
suNo,quizNo = map(int, input().split())
numbers = list(map(int, input().split()))
prifix_sum = [0]
temp = 0
for i in numbers:
    temp = temp + i
    prifix_sum.append(temp) #합 배열 만들기
for i in range(quizNo):
    s, e = map(int, input().split())
    print(prifix_sum[e]-prifix_sum[s-1]) #합 배열에서 구간합 구하기
```
