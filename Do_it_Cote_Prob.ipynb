```python
# 개념 : 구간 합
## 003 구간 합 구하기 1 (난이도 실버3, 백준 11659번)
# 입력1 : 데이터의 개수(1 ~ 100,000) & 질의 개수
# 입력2 : 구간(i,j) 합을 구할 대상 배열
# 출력 : i번째 수에서 j번째 수까지의 합

# 백준에서 시간초과 문제 해결
# import sys
# input = sys.stdin.readline

# 입력1 : 데이터의 개수(1 ~ 100,000) & 질의 개수
num_data, num_question  = map(int, input().split())

# 입력2 : 합을 구할 대상 배열
data = list(map(int, input().split()))

# 합 배열 구하기
section_sum = [0]
temp = 0
for i in data:
    temp += i
    section_sum.append(temp)
    # import pdb; pdb.set_trace()

# 구간(i,j) 입력 받아 구간 합 구하기
for i in range(num_question):
    start_idx, end_idx  = map(int, input().split())
    print(f"{section_sum[end_idx] - section_sum[start_idx-1]}")

del num_data, num_question, data, section_sum
```

```python
# 개념 : 구간 합
## 004 구간 합 구하기 2 (난이도 실버1, 백준 11660번)
# 입력1 : 표의 크기 N, 합을 구해야 하는 횟수 M
# 입력2 : N개의 줄에는 표에 채워져 있는 수가 1행부터 차례대로 주어진다
# 입력3 : M개의 줄에는 네 개의 정수 x1, y1, x2, y2
# 출력 : 총 M줄에 걸쳐 (x1, y1)부터 (x2, y2)까지 합

# 백준에서 시간초과 문제 해결
# import sys
# input = sys.stdin.readline

# N, M 입력받기 
N, M  = map(int, input().split()) 

# A : 1차원 리스트 
A = [[0] * (N + 1)]
# D : 2차원 리스트
D = [[0] * (N + 1) for _ in range(N + 1)] 

# A_row를 N차원 만큼 입력받아, 2차원 리스트 A 
for i in range(N):
    A_row = [0] + [int(x) for x in input().split()]
    A.append(A_row)

# 합 배열 D
for i in range(1, N+1):
    for j in range(1, N+1):
        D[i][j] = D[i][j-1] + D[i-1][j] - D[i-1][j-1] + A[i][j]

# 구간 합 result
for i in range(M):
    x1, y1, x2, y2 = map(int, input().split())
    result = D[x2][y2] - D[x1-1][y2] - D[x2][y1-1] + D[x1-1][y1-1]
    print(result)
```

```python
# 개념 : 투 포인터
## 006 연속된 자연수의 합 구하기 (난이도 실버5, 백준 2018번)
# 입력1 : 정수 N
# 출력 : 연속된 숫자의 합이 N이 되는 가지수

# 백준에서 시간초과 문제 해결
# import sys
# input = sys.stdin.readline

n = int(input())
count = 1
start_index = 1
end_index = 1
sum = 1

while end_index != n:
    if sum < n:
        end_index += 1
        sum += end_index
    elif sum == n:
        # print(f"start_index, end_index = {start_index, end_index}")
        end_index += 1
        sum += end_index
        count += 1
    elif sum > n:
        sum -= start_index
        start_index += 1
        #import pdb; pdb.set_trace()
        
# print(f"count = {count}")
print(count)
```

```python
# 개념 : 슬라이딩 윈도우
## 009 DNA 비밀번호 (난이도 실버5, 백준 12891번)
# 입력 1: 임의로 만든 DNA 문자열 길이 |S|, 비밀번호로 사용할 부분문자열의 길이 |P|  
# 입력 2: 임의로 만든 DNA 문자열
# 입력 3: 부분문자열에 포함되어야 할 {‘A’, ‘C’, ‘G’, ‘T’} 의 최소 개수가 공백을 구분으로 주어진다.
# 출력 : 만들 수 있는 비밀번호의 종류의 수

# 백준에서 시간초과 문제 해결
# import sys
# input = sys.stdin.readline

checkArr = [0] * 4
myArr = [0] * 4
checkSecret = 0

def myadd(c): #새로 들어온 문자를 처리하는 함수
    global checkArr,myArr,checkSecret
    
    import pdb; pdb.set_trace()
    if c == 'A':
        myArr[0] += 1
        if myArr[0] == checkArr[0]:
            checkSecret += 1
    elif c == 'C':
        myArr[1] += 1
        if myArr[1] == checkArr[1]:
            checkSecret += 1
    elif c == 'G':
        myArr[2] += 1
        if myArr[2] == checkArr[2]:
            checkSecret += 1
    elif c == 'T':
        myArr[3] += 1
        if myArr[3] == checkArr[3]:
            checkSecret += 1

def myremove(c): #제거되는 문자를 처리하는 함수
    global checkArr, myArr, checkSecret
    
    import pdb; pdb.set_trace()
    if c == 'A':
        if myArr[0] == checkArr[0]:
            checkSecret -= 1
        myArr[0] -= 1
    elif c == 'C':
        if myArr[1] == checkArr[1]:
            checkSecret -= 1
        myArr[1] -= 1
    elif c == 'G':
        if myArr[2] == checkArr[2]:
            checkSecret -= 1
        myArr[2] -= 1
    elif c == 'T':
        if myArr[3] == checkArr[3]:
            checkSecret -= 1
        myArr[3] -= 1

# 입력 1: 임의로 만든 DNA 문자열 길이 |S|, 비밀번호로 사용할 부분문자열의 길이 |P| 
print("입력 1: 임의로 만든 DNA 문자열 길이 |S|, 비밀번호로 사용할 부분문자열의 길이 |P| ")
S, P = map(int, input().split())
Result = 0

# 입력 2: 임의로 만든 DNA 문자열
print("입력 2: 임의로 만든 DNA 문자열")
A = list(input())

# 입력 3: 부분문자열에 포함되어야 할 {‘A’, ‘C’, ‘G’, ‘T’} 의 최소 개수가 공백을 구분으로 주어진다.
print("입력 3: 부분문자열에 포함되어야 할 {‘A’, ‘C’, ‘G’, ‘T’} 의 최소 개수가 공백을 구분으로 주어진다.")
checkArr = list(map(int, input().split()))

for i in range(4):
    if checkArr[i] == 0:
        checkSecret += 1

for i in range(P):  #초기 P 부분 문자열 처리 부분
    myadd(A[i])

if checkSecret == 4: #4자릿수와 관련된 크기가 모두 충족될 때 유효한 비밀번호
    Result += 1

for i in range(P, S):
    j = i - P
    myadd(A[i])
    myremove(A[j])
    if checkSecret == 4:
        Result += 1
print(Result)
```

```python
# 개념 : 스택
## 011 스택으로 수열 만들기(난이도 실버5, 백준 1874번)
# 입력 1:수열의 개수 n
# 입력 2-1: 첫번째 수
# 입력 2-2: 두번째 수
# 입력 2-n: n번째 수
# 출력 pop된 수열

# 백준에서 시간초과 문제 해결
# import sys
# input = sys.stdin.readline

print("수열 개수(N) 입력")
N = int(input())

print("수열 리스트(A) 초기화")
A = [0] * N

print("N개수 만큼 수열 입력")
for i in range(N):
    A[i] = int(input())
    # print(A)

stack = []
num = 1
result = True
answer = ""

print("N개수 만큼 반복")

for i in range(N):
    su = A[i]
    #현재 수열 값 >= 오름차순 자연수: 값이 같아질 때까지 push() 수행
    if su >= num:
        while su >= num:
            stack.append(num)
            num += 1
            answer += "+\n"
        stack.pop()
        answer += "-\n"
    else:
        n = stack.pop()
        #스택의 가장 위의 수가 만들어야 하는 수열의 수보다 크면 수열을 출력할 수 없음
        if n > su:
            print("NO")
            result = False
            break
        else:
            answer += "-\n"
if result:
    print(answer)



```

```python
# 개념 : 큐
## 013 카드게임(난이도 실버4, 백준 2164번)
# 입력 1: 카드 개수
# 출력 : 마지막 남는 카드

from collections import deque
N = int(input())
myQueue = deque()
for i in range(1, N+1):
    myQueue.append(i)
while len(myQueue) > 1:     # 카드가 1장 남을 때까지
    myQueue.popleft()       # 맨 위의 카드를 버림
    myQueue.append(myQueue.popleft())   # 맨 위의 카드를 가장 아래 카드 밑으로 이동
print(myQueue[0])   # 마지막으로 남은 카드 출력
```

```python
# 개념 : 버블정렬
## 015 수 정렬하기(난이도 브론즈1, 백준 2750번)

N = int(input())
A = [0]*N

for i in range(N):
    A[i] = int(input())

for i in range(N-1):
    for j in range(N-1-i):
        if A[j] > A[j+1]:
            temp = A[j]
            A[j] = A[j+1]
            A[j+1] = temp

for i in range(N):
    print(A[i])
```

```python
# 개념 : 선택정렬
## 017 내림차순으로 자릿수 정렬하기(난이도 실버5, 백준 1427번)

import sys
print = sys.stdout.write
A = list(input())
for i in range(len(A)):
    Max = i # 맨 앞에 있는 인덱스가 최대값의 인덱스로 가정
    for j in range(i+1, len(A)):
        if A[j] > A[Max]:   # 내림차순이므로 최댓값을 찾음
            Max = j
    if A[i] < A[Max]:
        temp = A[i]
        A[i] = A[Max]
        A[Max] = temp
for i in range(len(A)):
    print(A[i])
```

```python
# 개념 : 삽입정렬
## 018 ATM 인출 시간 계산하기(난이도 실버3, 백준 11399번)
N = int(input())
A = list(map(int, input().split()))
S = [0]*N
for i in range(1, N):    # 삽입 정렬
    insert_point = i #현재의 인덱스
    insert_value = A[i]
    for j in range(i-1, -1, -1):# 현재 앞칸 부터 -1 방향으로 0까지 간다
        if A[j] < A[i]:
            insert_point = j + 1
            break
        if j == 0:
            insert_point = 0
    for j in range(i, insert_point, -1):
        A[j] = A[j-1]
    A[insert_point] = insert_value
S[0] = A[0]
for i in range(1, N):    # 합 배열 만들기
    S[i] = S[i-1] + A[i]
sum = 0
for i in range(0, N):    # 합 배열 총합 구하기
    sum += S[i]
print(sum)
```
